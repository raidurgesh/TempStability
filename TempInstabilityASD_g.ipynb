```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00002, 1, 2000)
T_NoiseASD=0.1*((1+(2E-3/freq)**4))**0.5
plt.figure(figsize=(10,8))
plt.plot(freq, T_NoiseASD, 'r-',label = "3K/√Hz √(1+(0.1mHz/f)^4)")
plt.yscale("log")
plt.xscale("log")
plt.xlim(1e-5, 1)
plt.title("Temperature Instability ASD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Temperature Instability (K/√Hz)")
plt.legend(title = "ASD")
plt.plot(freq,np.ones(np.shape(freq))*3,'k-.')
plt.axvline(x = 0.00018)
plt.axvline(x = 0.00036)
plt.axvline(x = 0.0018)
plt.grid(visible=None, which='minor', axis='both')
plt.show()
```
