```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(1e-5, 1, 50000)
u_oms=((1+(2E-3/freq)**4))
u_therm=((1+((2*1.414)*1e-3/freq)**4))
u_KN=1e-3/freq
u_oms_ASD=u_oms**0.5
u_therm_ASD=u_therm**0.5
u_KN_ASD=u_KN**0.5
# PMON uses nominal power allocation = 100 ÂµW
PMON_stability_base=1e-4
# PMON Total dissipation
PMON_dissipation=PMON_stability_base*u_oms_ASD
PMON_dissipation_KN=PMON_stability_base*u_oms_ASD*u_KN_ASD
```

```python
plt.figure(figsize=(10,8))
plt.plot(freq, u_oms_ASD, 'r-',label = "â(1+(2mHz/f)^4)")
plt.plot(freq, u_therm_ASD, 'b-',label = "â(1+(2â2mHz/f)^4)")
plt.plot(freq, u_KN, 'g--',label = "1e-2/f")
plt.plot(freq, u_KN_ASD, 'g-',label = "â1e-2/f")
plt.plot(freq, PMON_dissipation, 'c-',label = "10ÂµK/âHz*â(1+(2mHz/f)^4): Old")
plt.plot(freq, PMON_dissipation_KN, 'k-',label = "10ÂµK/âHz*â(1+(2mHz/f)^4)*â1e-2/f")
plt.yscale("log")
plt.xscale("log")
plt.xlim(1e-5, 1)
plt.title("Thermal Stability, ASD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Temperature Instability (W/âHz)")
plt.legend()
plt.grid(visible=None, which='both', axis='both')
plt.show()
```
