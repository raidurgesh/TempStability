```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(1e-5, 1, 50000)
u_oms=((1+(2E-3/freq)**4))
u_therm=((1+((2*1.414)*1e-3/freq)**4))
# Added power law stability factor
u_KN=1e-3/freq
# Take root of all PSD form factors
u_oms_ASD=u_oms**0.5
u_therm_ASD=u_therm**0.5
u_KN_ASD=u_KN**0.5
# LH uses nominal power allocation = 145W, FRSo 1W, FRe 20W and PMON 0.1mW
LH_stability_base=145
FRSe_stability_base=20
FRSo_stability_base=1
PMON_stability_base=1e-4
Thermal_stability_base=0.1
# PMON Total dissipation
LH_dissipation=LH_stability_base*u_oms_ASD
LH_dissipation_KN=LH_stability_base*u_oms_ASD*u_KN_ASD
FRSe_dissipation=FRSe_stability_base*u_oms_ASD
FRSe_dissipation_KN=FRSe_stability_base*u_oms_ASD*u_KN_ASD
FRSo_dissipation=FRSo_stability_base*u_oms_ASD
FRSo_dissipation_KN=FRSo_stability_base*u_oms_ASD*u_KN_ASD
PMON_dissipation=PMON_stability_base*u_oms_ASD
PMON_dissipation_KN=PMON_stability_base*u_oms_ASD*u_KN_ASD
Thermal_stability=Thermal_stability_base*u_oms_ASD
Thermal_stability_KN=Thermal_stability_base*u_oms_ASD*u_KN_ASD
PMON_dissipation_rad=0.05*PMON_stability_base*u_oms_ASD
PMON_dissipation_KN_rad=0.05*PMON_stability_base*u_oms_ASD*u_KN_ASD
PMON_dissipation_cond=0.95*PMON_stability_base*u_oms_ASD
PMON_dissipation_KN_cond=0.95*PMON_stability_base*u_oms_ASD*u_KN_ASD
```

```python
plt.figure(figsize=(10,8))
plt.plot(freq, u_oms_ASD, 'r-',label = "√(1+(2mHz/f)^4): OMS")
plt.plot(freq, u_therm_ASD, 'b-',label = "√(1+(2√2mHz/f)^4: Thermal)")
plt.plot(freq, u_KN, 'g--',label = "1mHz/f")
plt.plot(freq, u_KN_ASD, 'g-',label = "√1mHz/f: KN Modification")
plt.plot(freq, PMON_dissipation, 'c-',label = "100µW/√Hz*√(1+(2mHz/f)^4): Old")
plt.plot(freq, PMON_dissipation_KN, 'k-',label = "100µW/√Hz*√(1+(2mHz/f)^4)*√1mHz/f: New")
plt.plot(freq,np.ones(np.shape(freq))*1e-4,'c-.', label = "100µW Baseline")
plt.yscale("log")
plt.xscale("log")
plt.xlim(1e-5, 1)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Thermal & Radiative Power Stability (W/√Hz)")
plt.legend()
plt.grid(visible=None, which='both', axis='both')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(freq, PMON_dissipation, 'c-',label = "100µW/√Hz*√(1+(2mHz/f)^4): Old")
plt.plot(freq, u_KN_ASD, 'g-',label = "√1mHz/f: KN Modification")
plt.plot(freq, PMON_dissipation_KN, 'k-',label = "100µW/√Hz*√(1+(2mHz/f)^4)*√1mHz/f: New")
plt.plot(freq,np.ones(np.shape(freq))*1e-4,'c-.', label = "100µW Baseline")
plt.title("PMON Power Instability, ASD")
plt.yscale("log")
plt.xscale("log")
plt.xlim(1e-5, 1)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Thermal & Radiative Power Stability (W/√Hz)")
plt.legend()
plt.grid(visible=None, which='both', axis='both')
plt.show()
```
