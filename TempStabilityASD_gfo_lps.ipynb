```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 50000)
T_NoiseASD=0.1*((1+(2E-3/freq)**4))**0.5
T_NoiseASD3=3*((1+(1E-4/freq)**4))**0.5
T_NoiseASD_sim=0.1*((1+(1E-4/freq)**4))**0.5
T_NoiseASD_p18=0.1*((1+(2E-3/0.18e-3)**4))**0.5
T_NoiseASD_p36=0.1*((1+(2E-3/0.36e-3)**4))**0.5
T_NoiseASD_1p8=0.1*((1+(2E-3/1.8e-3)**4))**0.5
```

```python
plt.figure(figsize=(10,8))
plt.plot(freq, T_NoiseASD, 'r-',label = "0.1˚K/√Hz √(1+(2mHz/f)^4)")
plt.plot(freq, T_NoiseASD3, 'g-',label = "3˚K/√Hz √(1+(0.1mHz/f)^4)")
plt.yscale("log")
plt.xscale("log")
plt.xlim(1e-5, 1)
plt.title("Temperature Instability, ASD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Temperature Instability (˚K/√Hz)")
plt.legend()
plt.plot(freq,np.ones(np.shape(freq))*3,'k-.')
plt.axvline(x = 0.00018)
plt.axvline(x = 0.00036)
plt.axvline(x = 0.0018)
plt.grid(visible=None, which='minor', axis='both')
plt.show()
```
